:class: important
Докажите равенство 
$\underbrace{0.00\ldots 0}_{n\text{ нулей}} (9) = \underbrace{0.00\ldots 0}_{n - 1\text{ нуль}}1.$

```{admonition} Решение
:class: tip, dropdown
Имеем
    
$$
    \underbrace{0.00\ldots 0}_{n\text{ нулей}} (9) = \frac 9{10^n} + \frac9{10^{n+1}} + \ldots = 
    \frac 9{10^n}\Big(1 + \frac 1{10} + \frac 1{100} + \ldots\Big).
$$
    
В скобках находится сумма геометрической прогрессии со знаменателем $q=\frac 1{10}$. По известной формуле находим, что эта сумма равна $\frac 1{1-q} = \frac{10}9$. Таким образом,

$$
    \underbrace{0.00\ldots 0}_{n\text{ нулей}} (9) = \frac 9{10^n} \cdot \frac{10}9 = \frac 1{10^{n-1}} = \underbrace{0.00\ldots 0}_{n - 1\text{ нуль}}1.
$$  
```

```{admonition} Монотонные последовательности
:class: note
Последовательность $(x_n)$ 
* **возрастающая**, если $x_{n+1} > x_n$ при всех $n \in\mathbb N$;
* **неубывающая**, если $x_{n+1} \geqslant x_n$ при всех $n \in\mathbb N$;
* **убывающая**, если $x_{n+1} < x_n$ при всех $n \in\mathbb N$;
* **невозрастающая**, если $x_{n+1} \leqslant x_n$ при всех $n \in\mathbb N$.

Если последовательность $(x_n)$ обладает хотя бы одним из перечисленных свойств, то она называется **монотонной**.
```

```{admonition} Ограниченные последовательности
:class: note
Последовательность $(x_n)$ 
* **ограничена сверху**, если существует такое число $M$, что $x_n < M$ при всех $n \in\mathbb N$;
* **ограничена снизу**, если существует такое число $m$, что $x_n >m$ при всех $n \in\mathbb N$;
* **ограничена**, если существует такое число $C$, что $\vert x_n\vert < C$ при всех $n \in\mathbb N$.
```
